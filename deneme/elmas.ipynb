```python
import seaborn as sns
import numpy as np
import pandas as pd
import time


class DiamondStats:
   def __init__(self):
       self.diamonds = sns.load_dataset('diamonds')
       self.cut_order = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
       self.color_order = ['J', 'I', 'H', 'G', 'F', 'E', 'D']
       self.clarity_order = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']


   def ozetDataset(self):
       ozet = self.diamonds.describe().T
       return ozet


   def kesimeGoreOrtUcret(self):
       print("Lütfen bir kesim türü seçiniz:\n1. Fair\n2. Good\n3. Very Good\n4. Premium\n5. Ideal")
       kesim=input("Seçiminiz:")
       if kesim=="1":
            kesime_gore_ort_ucret = self.diamonds.groupby('cut').mean()['price']['Fair']
            print("Ortalama ücret: ",kesime_gore_ort_ucret)
       elif kesim=="2":
           kesime_gore_ort_ucret = self.diamonds.groupby('cut').mean()['price']['Good']
           print("Ortalama ücret: ",kesime_gore_ort_ucret)
       elif kesim=="3":
           kesime_gore_ort_ucret = self.diamonds.groupby('cut').mean()['price']['Very Good']
           print("Ortalama ücret: ",kesime_gore_ort_ucret)
       elif kesim=="4":
           kesime_gore_ort_ucret = self.diamonds.groupby('cut').mean()['price']['Premium']
           print("Ortalama ücret: ",kesime_gore_ort_ucret)
       elif kesim=="5":
           kesime_gore_ort_ucret = self.diamonds.groupby('cut').mean()['price']['Ideal']
           print("Ortalama ücret: ",kesime_gore_ort_ucret)
       else:
           print("Geçersiz bir giriş yaptınız.")
           


   def rengeGoreOrtUcret(self):
       print("Lütfen bir renk seçiniz:\n1. J\n2. I\n3. H\n4. G\n5. F\n6. E\n7. D")
       renk=input("Seçiminiz: ")
       if renk=="1":
            renge_gore_ort_ucret = self.diamonds.groupby('color').mean()['price']['J']
            print("Ortalama ücret: ",renge_gore_ort_ucret)
       elif renk=="2":
           renge_gore_ort_ucret = self.diamonds.groupby('color').mean()['price']['I']
           print("Ortalama ücret: ",renge_gore_ort_ucret)
       elif renk=="3":
           renge_gore_ort_ucret = self.diamonds.groupby('color').mean()['price']['H']
           print("Ortalama ücret: ",renge_gore_ort_ucret)
       elif renk=="4":
           renge_gore_ort_ucret = self.diamonds.groupby('color').mean()['price']['G']
           print("Ortalama ücret: ",renge_gore_ort_ucret)
       elif renk=="5":
           renge_gore_ort_ucret = self.diamonds.groupby('color').mean()['price']['F']
           print("Ortalama ücret: ",renge_gore_ort_ucret)
       elif renk=="6":
           renge_gore_ort_ucret = self.diamonds.groupby('color').mean()['price']['E']
           print("Ortalama ücret: ",renge_gore_ort_ucret)
       elif renk=="7":
           renge_gore_ort_ucret = self.diamonds.groupby('color').mean()['price']['D']
           print("Ortalama ücret: ",renge_gore_ort_ucret)
       else:
           print("Geçersiz bir giriş yaptınız.")
 

   def berraklıgaGoreOrtUcret(self):
       print("Lütfen bir kesim türü seçiniz:\n1. I1\n2. SI2\n3. SI1\n4. VS2\n5. VVS2\n6. VVS1\n7. IF")
       berraklık=input("Seçiniz:")
       if berraklık=="1":
            berraklıga_gore_ort_ucret = self.diamonds.groupby('clarity').mean()['price']['I1']
            print("Ortalama ücret: ",berraklıga_gore_ort_ucret)
       elif berraklık=="2":
           berraklıga_gore_ort_ucret = self.diamonds.groupby('clarity').mean()['price']['SI2']
           print("Ortalama ücret: ",berraklıga_gore_ort_ucret)
       elif berraklık=="3":
           berraklıga_gore_ort_ucret = self.diamonds.groupby('clarity').mean()['price']['SI1']
           print("Ortalama ücret: ",berraklıga_gore_ort_ucret)
       elif berraklık=="4":
           berraklıga_gore_ort_ucret = self.diamonds.groupby('clarity').mean()['price']['VS2']
           print("Ortalama ücret: ",berraklıga_gore_ort_ucret)
       elif berraklık=="5":
           berraklıga_gore_ort_ucret = self.diamonds.groupby('clarity').mean()['price']['VVS2']
           print("Ortalama ücret: ",berraklıga_gore_ort_ucret)
       elif berraklık=="6":
           berraklıga_gore_ort_ucret = self.diamonds.groupby('clarity').mean()['price']['VVS1']
           print("Ortalama ücret: ",berraklıga_gore_ort_ucret)
       elif berraklık=="7":
           berraklıga_gore_ort_ucret = self.diamonds.groupby('clarity').mean()['price']['IF']
           print("Ortalama ücret: ",berraklıga_gore_ort_ucret)
       else:
           print("Geçersiz bir giriş yaptınız.")
       
def main():
   ds = DiamondStats()
   print("Pırlanta Dünyasına Hoşgeldiniz!")
   while True:
       print("Lütfen yapmak için bir işlem seçin:")
       print("1. Pırlanta istatistikleri")
       print("2. Kesime göre ücret")
       print("3. Renge göre ücret")
       print("4. Berraklık derecesine göre ücret")
       print("5. Çıkış")
       try:
           option = int(input("Option: "))
           if option == 1:
               summary = ds.ozetDataset()
               print(summary)
           elif option == 2:
               mean_price_by_cut = ds.kesimeGoreOrtUcret()
               print(mean_price_by_cut)
           elif option == 3:
               mean_price_by_color = ds.rengeGoreOrtUcret()
               print(mean_price_by_color)
           elif option == 4:
               mean_price_by_clarity = ds.berraklıgaGoreOrtUcret()
               print(mean_price_by_clarity)
           elif option == 5:
               print("...Çıkış yapılıyor...")
               time.sleep(1)
               print("Çıkış yapıldı.")
               break
           else:
               print("Geçerli olmayan işlem seçtiniz!")
       except ValueError:
           print("Geçerli olmayan giriş yaptınız! Lütfen bir sayı giriniz.")
           continue


if __name__ == '__main__':
   main()
```
