```python
import requests, time, socket, sys
from bs4 import BeautifulSoup
import datetime
# Şu anki tarih ve saat bilgisi
bugun = datetime.datetime.now()
# Tarih formatını ayarlayalım
tarih_formati = "{}.{}.{}".format(bugun.day, bugun.month, bugun.year)
# Saat formatını ayarlayalım
saat_formati = "{}:{}".format(bugun.hour, bugun.minute)
# 925 ayar Gümüşün fiyatını İnternet Sitesinden Çekiyoruz.
url = "https://bigpara.hurriyet.com.tr/kobi/dunya-emtia-borsalari/sxaggr-gumus-gram/"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
gram_gümüs = soup.find("span", {"class": "value dw"})
gram_gümüs = gram_gümüs.text.strip().replace('.','').replace(',','.')
gram_gümüs = float(gram_gümüs)
külce_gümüs = gram_gümüs + (gram_gümüs * (17/100))
sikke_gram = gram_gümüs + (gram_gümüs * (60/100))
a_gümüs_925 = gram_gümüs * (92.5/100)
a_gümüs_900 = gram_gümüs * (90/100)
a_gümüs_800 = gram_gümüs * (80/100)

while True:
    print("\n")
    print("*****************************")
    print("YetGen Kuyumculuk Sistemi'ne Hoş Geldiniz")
    print("*****************************\n")
    print("Lütfen yapmak için bir işlem seçin:")
    print("1. Güncel Gümüş Fiyatları\n2. Hazır Gümüş: Külçe ve Sikke\n3. Ziynet Eşyası Ör: Kolye,Bilezik,Küpe,Yüzük\n4. Hurda-Hasarlı Gümüş\n0. Çıkış")
    print("\n")
    try:
        option = int(input("Seçiminiz: "))
        if option == 1: #güncel gümüş fiyatları
                print("\n")
                print("{} tarihi, saat {} itibariyla güncel gümüş fiyatları:".format(tarih_formati, saat_formati))
                print("\n")
                print(f"Gram gümüş: {gram_gümüs}TL")
                print(f"Külçe gümüş: {külce_gümüs}TL")
                print(f"Sikke gümüş(işçiliksiz): {sikke_gram}TL")
                print(f"925 Ayar Gümüş: {a_gümüs_925}TL")
                print(f"900 Ayar Gümüş: {a_gümüs_900}TL")
                print(f"800 Ayar Gümüş: {a_gümüs_800}TL")
            
        elif option == 2: 
            print("Elinde Hangi Gümüşten Var?")
            print("1. Külçe Gümüş\n2. Sikke Gümüş")
            hazir_gümüs_secim = int(input("Seçiminiz: "))
            if hazir_gümüs_secim == 1:
                adet = int(input("Lütfen kaç adet külçe gümüş satmak istediğinizi girin: "))
                gram = float(input("Lütfen satılacak külçenin gram miktarını giriniz: "))
                toplam_fiyat = adet * gram * külce_gümüs
                print(f"{adet} külçe gümüş için sana biçtiğim toplam fiyat: {toplam_fiyat}TL")        
            elif hazir_gümüs_secim == 2:
                print("Lütfen satılacak sikke gümüşün işçilik seviyesini seçin:")
                print("1- İşçiliksiz")
                print("2- Kötü işçilik")
                print("3- Orta işçilik")
                print("4- İyi işçilik")
                sikke_ayari_secim = int(input("Seçiminiz: "))
                adet = int(input("Lütfen kaç adet sikke gümüş satmak istediğinizi girin: "))
                gram = float(input("Lütfen satılacak sikkenin gram miktarını giriniz: "))
                if sikke_ayari_secim == 1:
                    fiyat = adet * sikke_gram * gram
                    print(f"Toplam satış fiyatı: {fiyat} TL")
                elif sikke_ayari_secim == 2:
                    fiyat = adet * gram_gümüs * gram * 2
                    print(f"Toplam satış fiyatı: {fiyat} TL")
                elif sikke_ayari_secim == 3:
                    fiyat = adet * gram_gümüs * gram * 3
                    print(f"Toplam satış fiyatı: {fiyat} TL")
                elif sikke_ayari_secim == 4:
                    fiyat = adet * gram_gümüs * gram * 3.5
                    print(f"Toplam satış fiyatı: {fiyat} TL")
                else:
                    print("Geçersiz bir seçim yaptınız!")  
                    continue
                
        elif option == 3: 
                gümüs_turleri = ["925 ayar", "900 ayar", "800 ayar"]
                iscilik_turleri = ["kötü", "orta", "iyi", "çok iyi", "mükemmel"]
                print("Elinizdeki ziynet eşyasının ayarını seçin:")
                for index, gümüs in enumerate(gümüs_turleri, start=1):
                    print(f"{index}. {gümüs}")
                gümüs_secim = int(input("Seçiminizi yapınız: "))
                while gümüs_secim not in range(1, len(gümüs_turleri) + 1):
                    gümüs_secim = int(input("Geçersiz seçim. Lütfen tekrar deneyiniz: "))
                secilen_gümüs = gümüs_turleri[gümüs_secim - 1]
                gram = float(input("Eşyanın gram değerini giriniz: "))
                print("Aşağıdaki işçilik türlerinden birini seçiniz:")
                for index, iscilik in enumerate(iscilik_turleri, start=1):
                    print(f"{index}. {iscilik}")
                iscilik_secim = int(input("Seçiminizi yapınız: "))
                while iscilik_secim not in range(1, len(iscilik_turleri) + 1):
                    iscilik_secim = int(input("Geçersiz seçim. Lütfen tekrar deneyiniz: "))
                secilen_iscilik = iscilik_turleri[iscilik_secim - 1]
                gümüs_fiyat = {
                "925 ayar": a_gümüs_925,
                "900 ayar": a_gümüs_900,
                "800 ayar": a_gümüs_800
                }
                iscilik_oran = {
                    "kötü": 0.5,
                    "orta": 1,
                    "iyi": 3,
                    "çok iyi": 4,
                    "mükemmel": 5
                }
                fiyat = (float(gümüs_fiyat[secilen_gümüs]) * gram) * (1 + iscilik_oran[secilen_iscilik])
                print(f"Ürünün fiyatı: {fiyat:.2f} TL")
            
        elif option == 4: #hurda gümüşler gram fiyatına göre %20 aşağı satılır.
                print("Lütfen satılacak hasarlı gümüşün ayarını giriniz:")
                print("1- 1000 Ayar")
                print("2- 925 Ayar")
                print("3- 900 Ayar")
                print("4- 800 Ayar")
                gümüs_ayari_secim = int(input("Seçiminiz: "))
                gram = float(input("Lütfen satılacak ziynet eşyasının gram miktarını giriniz: "))
                if gümüs_ayari_secim == 1:
                    fiyat = gram_gümüs * gram * 0.8
                    print(f"Toplam satış fiyatı: {fiyat} TL")
                elif gümüs_ayari_secim == 2:
                    fiyat = a_gümüs_925 * gram * 0.8
                    print(f"Toplam satış fiyatı: {fiyat} TL")
                elif gümüs_ayari_secim == 3:
                    fiyat = a_gümüs_900 * gram * 0.8
                    print(f"Toplam satış fiyatı: {fiyat} TL")
                elif gümüs_ayari_secim == 4:
                    fiyat = a_gümüs_800 * gram * 0.8
                    print(f"Toplam satış fiyatı: {fiyat} TL")
                    
                else:
                    print("Geçersiz bir seçim yaptınız!")           
                    
        elif option == 0:
                print("...Program Kapanıyor...")
                time.sleep(2)
                print("Program kapandı.")
                break
        else:
                print("Geçerli olmayan işlem seçtiniz!")
        
    except ValueError:
        print("Geçerli olmayan giriş yaptınız! Lütfen bir sayı giriniz.")
        
```
